```python
%reload_ext autoreload
%autoreload 2

import datetime as dt
import sys
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append('..\..')

from Crypto.lib.backtest_statistics import account_curve_statistics
from Crypto.lib.constants import COINBASE_CRYPTOS

pd.options.mode.chained_assignment = None
```

```python
def strategy_trade_on_day(
    close: pd.DataFrame,
    days: list, 
    notional: float = 100,
    vol: pd.DataFrame = None,
    assets: list = None,
    start_date: dt.datetime = dt.datetime.min
) -> pd.DataFrame:
    """ Trade only on one day of the week. 0 for monday, 6 for sunday """
  
    position = close.copy()
    position = position[position.index >= start_date]
    if assets is not None:
        position = position[[col for col in position.columns if col in assets]]
    
    # Use spot of midnight on t for the full t+1 day
    position = position.resample("D", label='right').last().reindex(position.index, method="ffill")
    
    # Hold positions only on given days
    if isinstance(days, int):
        days = [days]
        
    position["day_of_week"] = position.index.dayofweek
    position = position.where(position["day_of_week"].isin(days), np.nan)
    position.drop("day_of_week", axis=1, inplace=True)
    
    # Construct weights
    nbr_assets = position.notnull().sum(axis=1)
    weights = (1. / position).div(nbr_assets, axis=0) * notional
    
    if vol is not None:
        weights = weights / vol
    
    return weights

def compute_daily_pnl(weights, close):
    notional = weights * close
    trades = weights.fillna(0.0).diff().fillna(0.0)
    notional_traded = (trades * close).resample("D").sum().abs().cumsum()
    
    pnl = (weights * close.diff()).fillna(0.)
    pnl_daily = pnl.resample("D").sum().cumsum()
    
    pnl_total = pnl.sum(axis=1)
    pnl_total_daily = pnl_daily.sum(axis=1)
    
    return pnl_daily, pnl_total_daily, notional, notional_traded
```

# Strategy PNL based on momentum

```python
pnl_daily, pnl_daily_total, notional, notional_traded = compute_daily_pnl(strategy_trade_on_day(close, days=[4, 5], assets=COINBASE_CRYPTOS, start_date=start_date), close)
```

```python
def get_relationship_df(pnl_daily, close, window=5):
    pnl = pnl_daily.diff().replace(0., np.nan)
    pnl = pnl.stack().to_frame("pnl")

    momentum = close.resample("d").last().pct_change(periods=window).shift()
    momentum = momentum.stack().to_frame("momentum")
    momentum = momentum.clip(momentum.momentum.quantile(0.01), momentum.momentum.quantile(0.99))

    res = pd.concat([pnl, momentum], axis=1).dropna(axis=0, how="any")
    return res

df = get_relationship_df(pnl_daily, close, window=5)
```

```python
sns.jointplot(data=df, x="momentum", y="pnl");
```

```python
def display_relationship(df, x, y, axe, n_quantiles=10):
    dft = df[[x, y]].copy()
    dft[f'{x}_quantile'] = pd.qcut(dft[x], q=n_quantiles, duplicates='drop', precision=2)

    dft.groupby(f"{x}_quantile").pnl.median().plot.bar(ax=axe)
    axe.set_xlabel(f"{x} quantile bucket")
    axe.set_ylabel(f"{y} median")
```

```python
fig, axes = plt.subplots(1, 5, figsize=(30, 5))
for i, window in enumerate([1, 2, 3, 4, 5]):
    df = get_relationship_df(pnl_daily, close, window=window)
    display_relationship(df, "momentum", "pnl", axes[i], n_quantiles=10)
    axes[i].set_title(f"pnl vs momentum over last {window} days")
```
